```python
from gensim import corpora, models, similarities
import pyLDAvis
import pyLDAvis.gensim
from sklearn.datasets import fetch_20newsgroups
pyLDAvis.enable_notebook()

dataset = fetch_20newsgroups()
corpus = dataset['data']
```

```python
tokenized_corpus = [text.split() for text in corpus]
dictionary = corpora.Dictionary(tokenized_corpus)

corpus_bow = [dictionary.doc2bow(text) for text in tokenized_corpus]
tfidf = models.TfidfModel(corpus_bow)

corpus_tfidf = tfidf[corpus_bow]

model = models.LdaModel(corpus_tfidf, id2word=dictionary, num_topics=5, passes=1)
corpus_model = model[corpus_tfidf]
```

```python

viz = pyLDAvis.gensim.prepare(model, corpus_bow, dictionary)
pyLDAvis.display(viz)
```
